# Map Coloring with heuristics

```python
# This methode returns the best state to color based on MRV    
def mrv(graph, visited_nodes, domains, domains_temp):
#   We retrieve all the nodes
    all_nodes = graph.keys()
#   We retrieve all the not-visited nodes
    not_visited_nodes =  list(set(all_nodes) - set(visited_nodes))
#   For each not-visited node we calculate its heuristic value
    node_hValue = {}
    for node in not_visited_nodes:
        node_hValue[len(domains_temp[node])] = node
    if len(list(node_hValue.keys())) == 0:
        return 'empty'
    else:
        return node_hValue[min(list(node_hValue.keys()))]

# This methode returns the best state to color based on DC
def dc(graph, visited_nodes):
#   We retrieve all the nodes
    all_nodes = graph.keys()
#   We retrieve all the not-visited nodes
    not_visited_nodes =  list(set(all_nodes) - set(visited_nodes))
    node_hValue = {}
    for node in not_visited_nodes:
        node_hValue[len(graph[node])] = node
    if len(list(node_hValue.keys())) == 0:
        return 'empty'
    else:
        return node_hValue[min(list(node_hValue.keys()))]

# This methode returns the best state to color based on LCV
def lcv(lcv_graph, lcv_visited_nodes, lcv_domains, lcv_domains_temp):
#   We retrieve all the nodes
    lcv_all_nodes = lcv_graph.keys()
#   We retrieve all the not-visited nodes
    lcv_not_visited_nodes =  list(set(lcv_all_nodes) - set(lcv_visited_nodes))
#   For each not-visited node we calculate its heuristic value
    lcv_node_hValue = {}
    for node in lcv_not_visited_nodes:
        lcv_node_hValue[len(lcv_domains_temp[node])] = node
    if len(list(lcv_node_hValue.keys())) == 0:
        return 'empty'
    else:
        return lcv_node_hValue[min(list(lcv_node_hValue.keys()))]
```

```python
def find_source(heuristic, visited_nodes, graph, domains_temp):
    if heuristic == 'mrv_heuristic':
        source = mrv(graph, visited_nodes, domains, domains_temp)
    if heuristic == 'dc_heuristic':
        source = dc(graph, visited_nodes)
    if heuristic == 'lcv_heuristic':
        source = lcv(graph, visited_nodes, domains, domains_temp)   
    return source
```

```python
def backTracking_fc_with_heuristics(graph, domains, source, length, isHeuristic, heuristicValue, domains_temp):
    visited_nodes   =[]
    assigned_colors ={}
    
    
    backtrack = 0
    while_loop = 0
    while True:
        print('source', source)
        
        #If all states are traversed then break the while
        if len(visited_nodes) == length:
            print('Assigned colors:', assigned_colors)
            print('Backtrack--->', backtrack)
            break
            
        #If source not in Visitied Nodes
        if source not in visited_nodes:
            
            #if colors are present in source domains
            if len(domains[source]) != 0:
                                
                #Assign color to source and delete from the domain
                assigned_colors[source] = domains[source][0]
                del domains[source][0]

                #After Assigning color remove the color from neighbor domains
                for neighbor in graph[source]:
                    if neighbor not in visited_nodes:
                        if assigned_colors[source] in domains[neighbor]:
                            #print(neighbor + '->' + assigned_colors[source] + ' removed')
#                            print('removing colors -->',assigned_colors[source])
                            domains[neighbor].remove(assigned_colors[source])
                    
                #Add source node to visited node
                print('Assigned color->', assigned_colors[source])
                visited_nodes.append(source)

                willContinue = False
                if isHeuristic:
                    #Take the next source using the approprate heuristic
                    source = find_source(heuristicValue, visited_nodes, graph, domains_temp)
                    if source == 'empty':
                        continue
                else:
                    #Take the next source neighbor if the neighbour is not presented in visited nodes
                    for neighbor in graph[source]:
                        if neighbor not in visited_nodes:
                            source = neighbor
                            willContinue = True
                            break
                        
                # When all neighbors are in visited nodes then go and check the remaining parent neighbours until
                # we find the neighbor not presented in visited nodes
                willContinue = False
                for node in reversed(visited_nodes):
                    for neighbor in graph[node]:
                        if neighbor not in visited_nodes:
                            source = neighbor
                            willContinue = True
                            break
                            
                    if willContinue:
                        break
                
            #If there are no colors available in source domain    
            else:
                
                #Traverse from last node of visited nodes
                for node in reversed(visited_nodes):
                    
                    willBreak = False
                    #print('else: node - ',node)
                    
                    #Assign node color to all neighbors
                    for neighbor in graph[node]:
                        if neighbor not in visited_nodes:
                            flag = True
                            
                            
                            # if there is any child which is having the same color of the current parent then
                            # we will not add the color to neighbor domain
                            for child in graph[neighbor]:
                                if child in visited_nodes and assigned_colors[child] == assigned_colors[node] and node != child:
                                    flag = False
                            #****************************************      
                                    
                            if flag:
                                domains[neighbor].append(assigned_colors[node])
                            
                    # If the node is the parent of the source        
                    if node in graph[source]:
                        #Assign node to source and break
                        print('Parent node->', node)
                        source = node;
                        willBreak = True
                        
                    #If the node is not the parent, Assign color to the node and remove from assigned colors
                    else:
                        domains[node].append(assigned_colors[node])
                    #print('Domains->',node,'->',domains[node] )
                        
                    
                    backtrack = 0
                    assigned_colors.pop(node)   #Remove key from assigned_colors
                    visited_nodes.remove(node)  #Remove key from visited nodes
                    
                    
                    if willBreak:
                        break
                        
                    
                    
                #If backtracking goes upto source then algorithm stops
                if len(visited_nodes) == 0:
                    print('Sorry!')
                    break
```
